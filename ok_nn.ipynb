```python
import pandas as pd
import numpy as np
import tensorflow as tf
import tensorflow.keras.layers as layers
import tensorflow.keras.models as models
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Precision, Recall
from sklearn.model_selection import KFold
batch_size = 10
no_epochs = 10
splits = 10

csv_data = pd.read_csv("~/Downloads/tracks_norm_1hot.csv", encoding='latin-1')
csv_data
```

```python
X = csv_data.drop(["artist", "track", "plays", "liked"], axis=1).to_numpy()
plays = csv_data["plays"].to_numpy()
liked = csv_data["liked"].to_numpy()
y = liked
num_features = X.shape[1]
```

```python
kfold = KFold(n_splits=splits, shuffle=True)
```

```python
histories = []
scores = []
for train_index, test_index in kfold.split(X):
    model = models.Sequential()
    model.add(layers.Flatten())
    model.add(layers.Dense(100, activation='relu'))
    model.add(layers.Dense(100, activation='relu'))
    model.add(layers.Dense(100, activation='relu'))
    model.add(layers.Dense(100, activation='relu'))
    model.add(layers.Dense(100, activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    X_train = X[train_index]
    X_test = X[test_index]
    y_train = y[train_index]
    y_test = y[test_index]
    history = model.fit(X_train, y_train, 
                       batch_size=batch_size,
                       epochs=no_epochs,
                       validation_split=.2,
                       verbose=0)
    score = model.evaluate(X_test, y_test, return_dict=True)
    histories.append(history.history)
    scores.append(score)
    
```
